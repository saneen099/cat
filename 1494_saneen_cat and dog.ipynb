### CNN PROJECT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.transform import resize
import glob
from sklearn.utils import shuffle
import tensorflow as tf
from tensorflow import keras
```

```python
cat_tr = glob.glob('cat and dog/test_set/cats/*.jpg')
dog_tr = glob.glob('cat and dog/test_set/dogs/*.jpg')
cat_tes = glob.glob('cat and dog/training_set/cats/*.jpg')
dog_tes = glob.glob('cat and dog/training_set/dogs/*.jpg')
```

```python
print(f'number of pictures presented in cat is {len(cat_tr)}')

print(f'number of pictures presented in dog is {len(dog_tr)}')

print(f'number of pictures for training in cat is {len(cat_tes)}')

print(f'number of pictures for training in dog is {len(dog_tes)}')
```

```python
image = []
label = []
for i in cat_tr:
    img =plt.imread(i)
    image.append(img)
    label.append(0)
for i in dog_tr:
    img = plt.imread(i)
    image.append(img)
    label.append(1)
```

```python
test_images = []
test_labels = []
for i in cat_tes:
    img = plt.imread(i)
    test_images.append(img)
    test_labels.append(0)

for i in dog_tes:
    img = plt.imread(i)
    test_images.append(img)
    test_labels.append(1) 
```

```python
image
```

```python
label
```

```python
image, label = shuffle(image,label)
```

```python
for i in range(20):
    plt.subplot(4,5,i+1)
    plt.imshow(image[i])
    plt.title(label[i])
    plt.tight_layout()
```

```python
xtrain = []
for i in image:
    xtrain.append(resize(i,(64,64,3)))
xtrain = np.array(xtrain)

xtest = []
for i in image:
    xtest.append(resize(i,(64,64,3)))
xtest = np.array(xtest)
```

```python
ytrain = np.array(label)
ytest = np.array(test_labels)
```

```python
ytrain
```

```python
ytest
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(16,(3,3),activation='elu',input_shape=(64,64,3)),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(32,(3,3),activation='elu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128,activation='elu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss= 'binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=30,validation_split=0.1)
```

```python
model.evaluate(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
pred
```

```python
xtest, pred = shuffle(xtest,pred)
```

```python
for i in range(20):
    #plt.figure(figsize=(10,12))
    plt.subplot(4,5,i+1)
    plt.imshow(xtest[i])
    if pred[i]>0.5:
        plt.title('YOU')
    else:
        plt.title('ME')
    plt.tight_layout()
```
